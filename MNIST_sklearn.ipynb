```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.datasets import load_digits
from sklearn.neighbors import KNeighborsClassifier
```

```python
digits=load_digits()
X=digits.data
y=digits.target
print('dimensiondeX:',X.shape)
```

```python
plt.imshow(digits['images'][0],cmap='Greys_r')#Train model
model=KNeighborsClassifier()
model.fit(X,y)
model.score(X,y)
```

```python
test=digits['images'][117].reshape(1,-1)
plt.imshow(digits['images'][117],cmap='Greys_r')
model.predict(test)
```
